```python
# So we need to scale this feature
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['Scaled_time'] = scaler.fit_transform(df['Time'].values.reshape(-1,1))

# Drop the original columns
df.drop(['Amount', 'Time'], axis=1, inplace=True)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Class distribution
print(df['Class'].value_counts())
# Expected: ~99.8% genuine, 0.2% fraud

# 2. Amount distribution
df[df['Class']==1]['Scaled_amount'].hist(bins=50, label='Fraud')
df[df['Class']==0]['Scaled_amount'].hist(bins=50, label='Genuine', alpha=0.5)
plt.legend()
plt.show()

# 3. Correlation heatmap (for top features)
sns.heatmap(df.corr(), cmap='coolwarm')
```
